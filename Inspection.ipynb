```python
import pickle
import pandas
from collections import Counter
import operator
```

```python
df_paths = [#'frames/mass_shootings_2013',
            #'frames/mass_shootings_2014',
            'frames/mass_shootings_2015']
frames = []
for df_path in df_paths:
    with open(df_path, 'rb') as infile:
        df = pickle.load(infile)
        frames.append(df)
df = pandas.concat(frames)
```

```python
print('number of killings')
len(df)
```

```python
states = Counter(row['state'] for index, row in df.iterrows())

counter = 0
for key, value in sorted(states.items(),
                         key=operator.itemgetter(1),
                         reverse=True):
    if counter < 10:
        print(key, value)
        counter += 1
    else:
        break
```

```python
num_sources = []
for index, row in df.iterrows():
    sources = row['incident_sources']
    sources.add(row['source_url'])
    num_sources.append(len(sources))

sources_distribution = Counter(num_sources)

counter = 0
for key, value in sorted(sources_distribution.items(),
                         key=operator.itemgetter(1),
                         reverse=True):
    if counter < 10:
        print(key, value)
        counter += 1
    else:
        break
```
